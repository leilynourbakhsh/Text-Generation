```python
def addCharFirst(text,char,place):
  return text[:place] + char + text[place:]
```

```python
def addCharEnd(text,char,place):
    return text[:place] + char + text[place:]
```

```python
class DataProcessor:

    def read_data(self,filename):
          with open (filename,encoding="utf8") as file:
              self.data = file.read().split('\n')
        
    def clean_text(self):
        
        clean2=[]
        for i in range(0,len(self.data)):
            new_res = re.sub('[\u06F0-\u06F9]+','N',self.data[i])
            clean2.append(new_res)
        clean1=[]
        for i in range(0,len(clean2)):
            clean2[i] = clean2[i].replace("؟"  ,    "adgg")
            clean2[i] = clean2[i].replace(".", "adgn")
            clean2[i]=re.sub(r'[^\w]',' ',clean2[i])
            clean2[i] = clean2[i].replace("adgg",   "؟")
            clean2[i] = clean2[i].replace("adgn",".")
            clean1.append(clean2[i])
        out3=[]
        for i in range(len(clean1)):
            clean1[i]=addCharFirst(clean1[i],' \s ',len(clean1[i])+1)
            clean1[i]=addCharEnd(clean1[i],' \e ',0)
            # print(clean1[i])
            out3.append(clean1[i])
        self.cleaned_data = out3
    def tokenizer(self):
        self.tokenized_data = [sub.split() for sub in self.cleaned_data]
        random.seed(87)
        random.shuffle(self.tokenized_data)
        train_size = int(len(self.tokenized_data) * 0.8)
        train_data = self.tokenized_data[0:train_size]
        self.test_data = self.tokenized_data[train_size:]
        self.tokenized_data = train_data

    def get_most_freq(self):
        frequences= {}

        for sentence in self.tokenized_data:
            for token in sentence:
                if token not in frequences.keys():
                    frequences[token] = 1
                else:
                    frequences[token]+=1
        #sort dictionaty  
        a=dict(sorted(frequences.items(), key=lambda item: item[1],reverse=True))
        self.n_tokens=sum(a.values())
        print(self.n_tokens)
        self.n_unique_tokens=len(a.values())
        print(self.n_unique_tokens)
        l_value=[]
        p=a.values()
        l_value = list(p)
        l_key=[]
        q=a.keys()
        l_key=list(q)
        self.frequences_dict= {l_key[i]:l_value[i] for i in range(len(l_key)-1000)}
    
    def handle_unknown(self):
        replace_sentences_tokens=[]
        vocab=[]
        vocab=list(self.frequences_dict.keys())
        

        for sentence in self.tokenized_data:
            replace_sentence=[]
            for token in sentence:
                if token in vocab:
                    replace_sentence.append(token)
                else:
                    replace_sentence.append('unk')

            replace_sentences_tokens.append(replace_sentence)
        self.prepared_data= replace_sentences_tokens
        

        
    def __init__(self,filename):
        self.read_data(filename)
        self.clean_text()
        self.tokenizer()
        self.get_most_freq()
        self.handle_unknown()
        for i in self.prepared_data:
            print(i)
        
```
